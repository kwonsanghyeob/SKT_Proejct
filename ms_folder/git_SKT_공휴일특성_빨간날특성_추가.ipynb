```python
import os
import sys
sys.path.append(os.pardir)

import numpy as np
import pandas as pd
from pytimekr import pytimekr
from utils import *
```

## EV 시계열 Load

```python
# EV 시계열 Load

filepath = 'data'
filename = "SKT_최종개별충전소_시계열_22.12.12"
print("[*] Load the dataset")

raw_df = pd.read_csv(os.path.join(filepath, "%s.csv" % filename),
                     engine='python', encoding='cp949',
                     index_col='datetime', parse_dates=True)       # index_col='수정 가능' !
raw_df.sort_index(inplace=True)

n_nulls = raw_df.isnull().sum().sum()
print(f"Null 값의 개수: {n_nulls}")
assert n_nulls == 0, "NaN 값 존재"

print("raw_df length :", len(raw_df))

ev_df = raw_df.copy()
```

## 공휴일 특성 추가
- 공휴일 의미 : 설날, 추석 등
- If 원핫 is False, Then, 공휴일이면 1 반환, 공휴일이 아니면 0 반환
- If 원핫 is True, Then, 공휴일이면 0,1 반환, 공휴일이 아니면 1,0 반환

```python
holidays = pytimekr.holidays(2020) + pytimekr.holidays(2021) + pytimekr.holidays(2022)

# 공휴일 특성 추가 (카테고리형)
holi_categ = holiday_feature(ev_df, holidays, is_onehot=False)

# shape
print(holi_categ.shape)

# 출력
display(pd.DataFrame(holi_categ))

# 공휴일 특성 추가 (원핫인코딩)
holi_onehot = holiday_feature(ev_df, holidays, is_onehot=True)

# shape
print(holi_onehot.shape)

# 출력
display(pd.DataFrame(holi_onehot).head(30))
```

## 빨간날 특성 추가
- 빨간날 의미 : 공휴일 + 토일
- If 원핫 is False, Then, 빨간날이면 1 반환, 빨간날이 아니면 0 반환
- If 원핫 is True, Then, 빨간날이면 0,1 반환, 빨간날이 아니면 1,0 반환

```python
holidays = pytimekr.holidays(2020) + pytimekr.holidays(2021) + pytimekr.holidays(2022)

# 공휴일 특성 추가 (카테고리형)
red_categ = redday_feature(ev_df, holidays, is_onehot=False)

# shape
print(red_categ.shape)

# 출력
display(pd.DataFrame(red_categ))

# 빨간날 특성 추가 (원핫인코딩)
red_onehot = redday_feature(ev_df, holidays, is_onehot=True)

# shape
print(red_onehot.shape)

# 출력
display(pd.DataFrame(red_onehot).head(30))
```
